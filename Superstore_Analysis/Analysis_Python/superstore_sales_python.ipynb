# EDA

## Categorical Analysis

```python
# Total Sales by Category:
Sales_by_Cat = round(df.groupby('Category')['Sales'].sum(),2)

print(Sales_by_Cat)
print()
print("Total Sales: ", df['Sales'].sum().round(2))
```

```python
# Total Profit by Category:
Profit_by_Cat = round(df.groupby('Category')['Profit'].sum(),2)

print(Profit_by_Cat)
print()
print("Total Profit:",df['Profit'].sum().round(2))
```

```python
# Sales by Sub-Category:
Sales_by_SubCat = round(df.groupby('Sub-Category')['Sales'].sum(),2)

print(Sales_by_SubCat)
print()
print("Total Sales:", df['Sales'].sum().round(2))
```

### Insights from Categorical Analysis
- Best-selling category: Technology
- Most profitable category: Technology
- Top-5 Loss-making sub-categories: Fasteners, Lables, Envelopes, Art, Supplies

## Regional Analysis

```python
# Sales by Region:
Sales_by_Reg = round(df.groupby('Region')['Sales'].sum(),2)

print(Sales_by_Reg)
print()
print("Total Sales: ", df['Sales'].sum().round(2))
```

```python
# Profit by Region:
Profit_by_Reg = round(df.groupby('Region')['Profit'].sum(),2)

print(Profit_by_Reg)
print()
print("Total Profit: ", df['Profit'].sum().round(2))
```

### Insights from Regional Analysis
- Highest sales region: West
- Lowest profit region: South

## Customer Segment Analysis

```python
# Sales by Segment
Sales_by_Seg = round(df.groupby('Segment')['Sales'].sum(),2)

print(Sales_by_Seg)
print()
print("Total Sales: ", df['Sales'].sum().round(2))
```

```python
# Profit by Segment
Profit_by_Seg = round(df.groupby('Segment')['Profit'].sum(),2)
print(Profit_by_Seg)
print()
print("Total Profit: ", df['Profit'].sum().round(2))
```

### Insights from Segment Analysis

```python
# Distribution
fig, axes = plt.subplots(2, 2, figsize=(12, 8))

# Sales
axes[0, 0].hist(df['Sales'], bins=30)
axes[0, 0].set_title('Sales Distribution')

# Profit
axes[0, 1].hist(df['Profit'], bins=30)
axes[0, 1].set_title('Profit Distribution')

# Quantity
axes[1, 0].hist(df['Quantity'], bins=30)
axes[1, 0].set_title('Quantity Distribution')

# Discount
axes[1, 1].hist(df['Discount'], bins=30)
axes[1, 1].set_title('Discount Distribution')

plt.tight_layout()
plt.show()
```

```python
# Distribution
fig, axes = plt.subplots(2, 2, figsize=(12, 8))

sns.boxplot(y=df['Sales'], ax=axes[0, 0])
axes[0, 0].set_title('Sales')

sns.boxplot(y=df['Profit'], ax=axes[0, 1])
axes[0, 1].set_title('Profit')

sns.boxplot(y=df['Quantity'], ax=axes[1, 0])
axes[1, 0].set_title('Quantity')

sns.boxplot(y=df['Discount'], ax=axes[1, 1])
axes[1, 1].set_title('Discount')

plt.tight_layout()
plt.show()
```

```python
# Discount vr profit
fig, axes = plt.subplots(2, 2, figsize=(12, 10))

# Sales vs Profit
sns.scatterplot(x='Sales', y='Profit', data=df, ax=axes[0,0])
axes[0,0].set_title('Sales vs Profit')

# Discount vs Profit
sns.scatterplot(x='Discount', y='Profit', data=df, ax=axes[0,1])
axes[0,1].set_title('Discount vs Profit')

# Quantity vs Profit
sns.scatterplot(x='Quantity', y='Profit', data=df, ax=axes[1,0])
axes[1,0].set_title('Quantity vs Profit')

# Sales vs Discount
sns.scatterplot(x='Sales', y='Discount', data=df, ax=axes[1,1])
axes[1,1].set_title('Sales vs Discount')

plt.tight_layout()
plt.show()
```

### Insights:
1. Sales vr Profit: 
- Positive relationship.
- Sales ^ - Profit ^
- Few high-sales orders result in loss --> some unprofitable transactions.
2. Discount vs Profit:
- High discount reduces Profit.
- Orders with 50% DIscount --> Low/Negative Profit.
3. Quantity vr Profit:
- Weak relationship.
- Selling more units doesn't always lead to higher profit; profitability depends on factors like discount, product category, and margin.
4. Sales vs Discount:
- Most orders have low sales and low discounts
######
##### Sales show a positive impact on profit, while discounts negatively affect profitability. 
##### Quantity alone is not a strong predictor of profit. 
##### Most sales occur with low discounts, suggesting that maintaining moderate discount levels helps preserve profit margins.

```python
corr = df[['Sales','Profit','Quantity','Discount']].corr()

sns.heatmap(corr,
            annot=True,
            cmap='coolwarm')
```

### Insights
- Sales have strongest +ve relationship with Profit (0.48).
- Discount negatively impacts Profit (-0.22).
- Quantity has weak relationship with Sales (0.20) --> no relationship with Discount (0.0086).
- Managing discounts effectively is important for maintaining profitability.
